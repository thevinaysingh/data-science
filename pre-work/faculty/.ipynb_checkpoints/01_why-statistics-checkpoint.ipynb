# 2. Statistics as a language of variation

Traditionally, software engineers have relied on unit testing to confirm that performance of code developed locally hold true even when the code is rolled out into production.

However, two aspects of ML model deployment make it very difficult to execute stress tests locally - variation in user behavior & variation in model performance on production data.

As a consequence, we need a formal language to describe variation.
