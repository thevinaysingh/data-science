### Q.4. Write a program which will take two lists list1 & list2 as input and it will return a list having all the numbers divisible by 4 which are present in the given two lists.
- For example list1 = [2,4,16,23] & list2 = [5,6,8,12] Expected Output: [4,16,8,12]

```python
list1 = [2,4,16,23]
list2 = [5,6,8,12]

new_list = list1 + list2
div_by_four = []
for number in new_list:
  if number % 4 == 0:
    div_by_four.append(number)

print(div_by_four)
```

### Q.1. Find the largest number from given 3 numbers using the if, elif, else conditional statements. Take the inputs from the user and store the result in the “largest'' variable.

```python
#num1 = 10
#num2 = 14
#num3 = 12

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

if (num1 >= num2) and (num1 >= num3):
   largest = num1
elif (num2 >= num1) and (num2 >= num3):
   largest = num2
else:
   largest = num3

print(f"The largest number is : {largest} ")
```

### Q.2. Write a function to calculate sum of all natural numbers upto 50(take input from user). If number is less than zero, print "enter a positive number", else print the sum. Use 'while' loop to iterate until number is zero.

```python
def sum_of_natural_num(number):
  number = int(input("Enter the number : "))

  if number < 0:
    print("Please enter a positive number")
  else:
    sum = 0
    # use while loop to iterate until zero
    while(number > 0):
      sum += number
      number -= 1
    print(f"The sum of all natural numbers till 50 is : {sum} ")

sum_of_natural_num(number)
```

### Q.4. Find the factorial of a number given by user. For the entered negative number, the program should print "factorial does not exist for negative numbers". For zero, it should print "factorial of zero is 1". (Hint: use 'for' loop as well as proper 'if elif' conditional statements)

```python
number = int(input("Enter a number: "))
factorial = 1

# use if..elif..else statements in order to check if the number is negative, positive or zero
if number < 0:
   print("Factorial does not exist for negative numbers")
elif number == 0:
   print("The factorial of zero is 1")
else:
   for i in range(1, number + 1):
       factorial = factorial*i
   print(f"The factorial of {number} is {factorial} ")
```

### Q.5. Generate a new list from the below given list which has elements divisible by 12 only using a single line code. (Hint : Use ONLY "lambda" function)

```python
new_list = [12, 65, 48, 108, 112, 120, 85]

# use filter(function, iterable) to get all numbers divisible by 12
# use lambda function to check the divisibilty condition
div_by_tweleve = list(filter(lambda x: (x % 12 == 0), new_list))

print(f"The list of numbers divisible by 12 is : {div_by_tweleve} ")
```
