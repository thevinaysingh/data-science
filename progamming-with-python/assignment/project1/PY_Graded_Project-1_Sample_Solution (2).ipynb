### Q5.  Write a program which will take two lists X & Y as input and it will return a list having all the odd numbers present in the given two lists.
Input list: X=[1,2,3,4,5] & Y=[5,6,7,8,9] 
Expected Output: [1,3,5,5,7,9]


```python
X=[1,2,3,4,5]
Y=[5,6,7,8,9]

newlist = X+Y
oddlist = []
for i in newlist:
  if i%2!=0:
    oddlist.append(i)

print(oddlist)
```

### Q6. Write a function that prints the integers from 1 to 15. But for multiples of three print "GREAT" instead of the number, and for the multiples of five print "LEARNING". For numbers which are multiples of both three and five print "GREATLEARNING".

```python
def great_learning():
  for i in range(1, 16):
    if i%3 == 0 and i%5==0:
      print("GREATLEARNING")
    elif i%3==0:
      print('GREAT')
    elif i%5==0:
      print("LEARNING")
    else:
      print(i)

great_learning()
```
