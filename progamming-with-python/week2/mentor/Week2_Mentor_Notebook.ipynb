Write a program to print all the unique values in the dictionary.

```python
list_dict=[{'apple':1},{'orange':2},{'pineapple':1},{'guava':3},{'strawberry':5}]

unique=set()
for di in list_dict:
    for j in di.values():
        unique.add(j)

print(unique)
```

### Question 17:
- 16.1 Write a function to check whether a string is palindrome or not
- 16.2 Check whether 'Raman' and 'Madam' are palindrome or not

```python
def palidrome(var):
    org = var
    temp = var
    rev_str = temp[::-1]
    if org == rev_str:
        return True
    else:
        return False 
```

```python
palidrome('ram')
```

```python
palidrome('madam')
```

### Example-1
Write a program to check whether a number is a perfect number. a perfect number is defined as the positive number in which sum of all positive divisors excluding the number itself is equal to that number.

ex: 6 the divisor of 6 is 1,2,3 and 1+2+3=6 28 the divisors of 28 is 1,2,4,7,14 and 1+2+4+7+14=28

```python
def is_perfect(num):
    if num<1:
        print("The number is not a perfect number")
    sum=0
    for i in range(1,num):
        if num%i==0:
            sum+=i
    if sum==num:
        print("The number is perfect")
    else:
        print("The number is not perfect")

is_perfect(18)
```

### Example-2
Write a program which will take two ranges rg1(i.e.2) and rg2(i.e.50) and it will return list of twin primes [(3, 5), (7, 41), (11, 43), (13, 17), (19, 29), (31,)].

NOTE: ORDER SHOULD NOT MATTER.

```python
def is_prime(num):
    for i in range(2,num):
        if num%i==0:
            return False
    return True

def gen_twin(rg1,rg2):
    temp1=[]
    temp2=[]
    final_li=[]
    for i in range(rg1,rg2):
        j=i+2
        if(is_prime(i) and is_prime(j)):
            temp1.append(i)
            temp2.append(j)
    for i in temp1:
         temp2.append(i)
    temp2.sort()
    twin_set=set(temp2)
    for i in twin_set:
        final_li.append(i)
    t=final_li
    final=[tuple(t[i:i+2]) for i in range(0,len(t),2)]
    return final
    

prime_twin=gen_twin(2,50)
      
prime_twin
```
