# Variables and Operators

Variable name is known as identifier. There are few rules that you have to follow while naming the variables in Python.

Variables are used to store data, they take memory space based on the type of value we assigning to them. Creating variables in Python is simple, you just have write the variable name on the left side of = and the value on the right side, as shown below. You do not have to explicitly mention the type of the variable, python infer the type based on the value we are assigning.

- The name of the variable cannot have special characters such as %, $, # etc, they can only have alphanumeric characters and underscore (A to Z, a to z, 0-9 or _ ).
- Variable name is case sensitive in Python which means num and NUM are two different variables in python.

### Problem statement 5 : Remove all control characters from a given string:
"Great\0\r Le\bar\tni\0ng\f"

```python
input_string = 'Great\0\r Le\bar\tni\0ng\f'
  
# printing original string
print(f"The original string is : {str(input_string)} ")
  
# using translate() and fromkeys() in order to escape all control characters
mappings =  dict.fromkeys(range(32))
output_string = input_string.translate(mappings)

print(f"String after removing control characters : {str(output_string)} ")
```

translate() returns a string that is modified string of given string according to given translation mappings.

### Problem statement 1 :  Calculate Probability density function(PDF) and Cummulative distribution function(CDF) for a given random number.

The probability distribution function or PDF computes the likelihood of a single point in the distribution and the cumulative distribution function or CDF computes the likelihood of an interval of the distribution.

#### PDF

```python
import scipy
x = 1.0
pdf = scipy.stats.norm.pdf(x, loc=0, scale=1)
pdf
```

#### CDF

```python
x = 1.0
y = 0.5
cdf = scipy.stats.norm.cdf(x, loc=0, scale=1) - scipy.stats.norm.cdf(y, loc=0, scale=1)
cdf
```

#### Lets take a simple case study in order to understand different python packages such as numpy, pandas, matplotlib, seaborn, scikit learn etc.

#### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Create dataframe using Pandas

```python
states_US = ["Texas", "Washington", "New York", "Hawaii", "Florida", "California"]
population = [39613467, 38737711, 32345411, 45323321, 20944987, 20299981]

# Create dataframe
df = pd.DataFrame({'States': states_US, 'Population': population})
df
```

#### We can calculate the mean, variance and percentiles of population feature using NumPy build in functions

```python
mean_population = round(np.mean(df["Population"]),3)
mean_population
```

```python
var_population = np.var(df["Population"])
var_population
```

```python
Q1, Q3 = np.percentile(a = df["Population"], q = [25,75])
print(f" 25th percentile of population is {Q1} ")
print(f" 75th percentile of population is {Q3} ")
```

#### We can also check these statistics of data using Pandas describe() function.

```python
df.describe()
```

#### Scikit Learn package

#### Most of the machine learning algorithms works with only numeric data. We can use LabelEncoder class from sklearn preprocessing library in order to generate unique labels for State feature from data.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["States"] = le.fit_transform(list(df["States"]))
df
```

#### Lets visualize this data using Matplotlib and Seaborn packages and find inferences

```python
plt.bar(x=df["States"],height=df["Population"])
plt.xlabel("States")
plt.ylabel("Population")
plt.show()
```

We can also visualize the same inference using seaborn library

```python
sns.barplot(x=df["States"],
            y=df["Population"],
            palette="deep")
plt.show()
```

#### lmplot() method  in seaborn library is used to draw a scatter plot onto a FacetGrid.

```python
sns.lmplot(x = "States", y = "Population", data = df, hue = "States")
```

```python
sns.displot(data = df, x = "Population", kde = True)
```
